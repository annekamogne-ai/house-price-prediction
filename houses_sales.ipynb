### 2- Data Wrangling

```python
df.drop(['Unnamed: 0','id'],axis=1,inplace=True)
df.describe()
```

```python
print(f"bedrooms missing values: {df['bedrooms'].isnull().sum()}\nbathrooms missing values: {df['bathrooms'].isnull().sum()}")
```

```python
bed_mean=df['bedrooms'].mean()
bad_mean=df['bathrooms'].mean()
df['bathrooms'].replace(np.nan,bad_mean,inplace=True)
df['bedrooms'].replace(np.nan,bed_mean,inplace=True)
df.info()
```

 # 4- Model Development

Let's fit a linear regression model to predict the <code>'price'</code> using the feature <code>'sqft_living'</code> and calculate the R^2.

```python
lr=LinearRegression()
x=df[['sqft_living']]
y=df[['price']]
lr.fit(x,y)
lr.score(x,y)
```

Let's fit a linear regression model to predict the <code>'price'</code> using the list of features.

```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"] 
x=df[features]
y=df[['price']]
lr.fit(x,y)
lr.score(x,y)
```

Pipeline with PolynomialFeatures

```python
input=[('scale',StandardScaler()),('poly',PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
my_pipe=Pipeline(input)
my_pipe.fit(x,y)
my_pipe.score(x,y)
```

### 5- Model Evaluation and Refinement

We will split the data into training and testing sets, then create and fit a Ridge regression object using the training data, set the regularization parameter to 0.1, and calculate the R^2 using the test data.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=1)
rr=Ridge(alpha=0.1)
rr.fit(x_train,y_train)
rr.score(x_test,y_test)
```

Let's perform a second order polynomial transform on both the training data and testing data. Fit a Ridge regression  and calculate the R^2 utilising the test data provided.

```python
pr=PolynomialFeatures(degree=2,include_bias=False)
x_train_pr=pr.fit_transform(x_train)
x_test_pr=pr.fit_transform(x_test)
rr.fit(x_train_pr,y_train)
rr.score(x_test_pr,y_test)
```
